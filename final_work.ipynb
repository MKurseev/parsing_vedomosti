__Зачем я создаю словарь?__


> Так как каждый раз кликать при нахождении статьи оч долго, я решил сохранять все `urls` с датами при прохождении (это легче всего делать в словаре). А потом прохожусь по словарю и забираю все нужные текста!

```python
dict_help = {} 
```

__Aнь, это единственное что для тебя!__

* `start_date` - тут ты ставишь начало отсчета. **y.m.d**
* `date_range` - насколько дней вперед считать. 

___
*Тут нужно менять циферки**

```python
year = 2018

start_date = datetime.date(year, 3, 2)
date_range = 91

dict_help = {} 
```

___

#### Тут показано, в каком диапазоне будет поиск :

```python
end_date = start_date + datetime.timedelta(days=date_range-1)

print('Первый образотанный день : ' + str(start_date))
print('Последний образотанный день : ' + str(end_date))
```

___

```python
urls_f = open(f'all_texts/urls_{year}.txt', 'a' , encoding='utf-8')
titles_f = open(f'all_texts/titles_{year}.txt', 'a', encoding='utf-8')

for i in range(date_range):
    
    # входные параметры
    date = start_date + datetime.timedelta(days=i)
    
    
    if (i+1) % 10 == 0:
        print(str(date) + ' обработано до этой даты')
  
    # search
    urls, titles = list_of_url_titles(date, markers_wild)
    
    # check if there is findings
    find_articles = len(urls) != 0
    
    if find_articles:
        
        urls_f.write(str(date) + '\n')
        titles_f.write(str(date) + '\n')
        
        dict_help[str(date)] = []
        
        for i in range(len(urls)):
            
            urls_f.write(urls[i] + '\n')
            titles_f.write(titles[i] + '\n')
            
            dict_help[str(date)].append(urls[i])
    
urls_f.close()
titles_f.close()
```

___
#### Это прост посмотреть, что в файлах лежит

```python
with open(f'all_texts/titles_{year}.txt', "r", encoding='utf-8') as f:
    titles = f.read()
    print(titles)
    
with open(f'all_texts/urls_{year}.txt', "r", encoding='utf-8') as f:
    urls = f.read()
    print(urls)
```

___

```python
texts_f = open(f'all_texts/texts_{year}.txt', 'a' , encoding='utf-8')
for dates, urls in dict_help.items():
    
    texts_f.write(str(dates) + '\n')
    
    for url in urls:
        text = copy_text_from_article(url)
        texts_f.write(text + '\n')
        
texts_f.close()
```
